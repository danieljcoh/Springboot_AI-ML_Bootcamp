```python
print(f"\nTotal Pokémon: {len(pokemon_df)}")
print(f"Unique Names: {pokemon_df['Name'].nunique()}") # Are there duplicates?
```

```python
# How many are "pure" single-type Pokémon?
pure_type_count = pokemon_df['Type 2'].isnull().sum()
print(f"Pokémon with only one type: {pure_type_count}")
```

```python
# Question 1: What is the distribution of Primary Types (Type 1)?
# We'd expect a bar chart here.
plt.figure(figsize=(12, 6))
sns.countplot(data=pokemon_df, x='Type 1', order=pokemon_df['Type 1'].value_counts().index, palette='Spectral')
plt.title('Distribution of Primary Pokémon Types (Type 1)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('1_type1_distribution.png')
print("Saved '1_type1_distribution.png'")
```

```python
# Question 2: How many are Legendary?
# A simple pie chart or count plot is perfect.
plt.figure(figsize=(7, 5))
pokemon_df['Legendary'].value_counts().plot(kind='pie', autopct='%.1f%%', colors=['#66b3ff','#ff9999'])
plt.title('Legendary vs. Non-Legendary Pokémon')
plt.ylabel('') # Hide the 'Legendary' y-label
plt.savefig('2_legendary_pie_chart.png')
print("Saved '2_legendary_pie_chart.png'")
```

```python
# Question 3: What is the distribution of 'Total' stats?
# A histogram or KDE plot will show us if stats are normally distributed
# or if there are multiple "groups" (e.g., weak, normal, strong).
plt.figure(figsize=(10, 6))
# sns.histplot(pokemon_df['Total'], kde=True, bins=30, color='mediumpurple') -- FIX
plt.title('Distribution of Total Stat Points')
plt.xlabel('Total Stats')
plt.ylabel('Count')
plt.savefig('3_total_stats_histogram.png')
print("Saved '3_total_stats_histogram.png'")
```
