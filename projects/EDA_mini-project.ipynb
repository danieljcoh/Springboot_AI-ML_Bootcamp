### Step 7: Drop rows with missing values.

```python
df = df.dropna()
print(df.isnull().sum().sum())
```

### Step 8: Create a new column for trip duration in minutes

```python
df["trip_duration_min"] = (df["tpep_dropoff_datetime"] - df["tpep_pickup_datetime"]).dt.total_seconds() / 60
```

### Step 9: Create new columns for pickup hour and day of week

```python
df["pickup_hour"] = df["tpep_pickup_datetime"].dt.hour
df["pickup_dayofweek"] = df["tpep_pickup_datetime"].dt.day_name()
```

### Step 10: Create a lineplot displaying the number of trips by pickup hour

```python
# Count trips by hour
trips_by_hour = df["pickup_hour"].value_counts().sort_index()

# Line plot
sns.lineplot(x=trips_by_hour.index, y=trips_by_hour.values)
plt.xlabel("Pickup Hour")
plt.ylabel("Number of Trips")
plt.title("Number of Trips by Pickup Hour")
plt.show()
```

### Step 11: Create a lineplot displaying the number of trips by pickup day

```python
# Count trips by day of week
trips_by_day = df["pickup_dayofweek"].value_counts().reindex(
    ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
)

# Line plot
sns.lineplot(x=trips_by_day.index, y=trips_by_day.values)
plt.xlabel("Day of Week")
plt.ylabel("Number of Trips")
plt.title("Number of Trips by Pickup Day")
plt.show()
```

### Step 12: Compute correlation matrix of numerical variables

```python
corr_matrix = df.corr(numeric_only=True)
print(corr_matrix)
```

### Step 13: Create a heatmap of the correlation matrix

```python
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Numerical Variables")
plt.show()
```

### Step 14: Create a scatter plot matrix of numerical variables. If memory issues try the df.sample method.

```python
# Sample to avoid memory issues (e.g., 10,000 rows)
sample_df = df.sample(n=10000, random_state=42)

# Scatter plot matrix
sns.pairplot(sample_df, diag_kind="kde")
```

### Step 15: Create a Seaborn countplot for PULocationID and DOLocationID. Only plot the top 15 categories by value counts.

```python
# Top 15 pickup and dropoff locations
top_pu = df["PULocationID"].value_counts().nlargest(15).index
top_do = df["DOLocationID"].value_counts().nlargest(15).index

# Filter and plot
plt.figure(figsize=(12, 5))
sns.countplot(data=df[df["PULocationID"].isin(top_pu)], x="PULocationID", order=top_pu)
plt.title("Top 15 Pickup Locations")
plt.show()

plt.figure(figsize=(12, 5))
sns.countplot(data=df[df["DOLocationID"].isin(top_do)], x="DOLocationID", order=top_do)
plt.title("Top 15 Dropoff Locations")
plt.show()
```

## Step 16: Create a box plot of total amount by payment type. Do you see anything odd?

```python
sns.boxplot(data=df, x="payment_type", y="total_amount")
plt.title("Total Amount by Payment Type")
plt.show()
```

### Step 17: Explore data distributions for 'fare_amount', 'trip_distance' and 'extra' using Seaborn's histplot. Sample the data if you run into memory issues.

```python
# Sample to avoid memory issues
sample_df = df.sample(n=10000, random_state=42)

cols = ["fare_amount", "trip_distance", "extra"]

for col in cols:
    plt.figure()
    sns.histplot(sample_df[col], kde=True, bins=50)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```
