```python
import requests
import json

# URL of your local API (ensure main.py is running in a terminal first!)
API_URL = "http://127.0.0.1:8000/predict"

def simulate_battle(id1, id2):
    payload = {
        "pokemon_1_id": id1,
        "pokemon_2_id": id2
    }
    
    try:
        response = requests.post(API_URL, json=payload)
        if response.status_code == 200:
            data = response.json()
            print(f"‚öîÔ∏è Battle: {data['p1_name']} vs {data['p2_name']}")
            print(f"üèÜ Predicted Winner: {data['winner']}")
            print(f"üìä Confidence: {data['win_probability']:.1%}")
        else:
            print("Error:", response.text)
    except requests.exceptions.ConnectionError:
        print("‚ùå Could not connect to API. Is 'main.py' running?")

# Test 1: Mewtwo (150) vs Caterpie (10)
simulate_battle(150, 10)

# Test 2: Charizard (6) vs Blastoise (9)
simulate_battle(6, 9)
```
